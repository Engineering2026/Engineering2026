```python
x = float(input('Enter Score:  '))
if 90<=x<=100:
  grade = 'A'
elif 80<=x<=89:
  grade = 'B'
elif 70<=x<=79:
  grade = 'C'
elif 60<=x<=69:
  grade = 'D'
elif 0<=x<=59:
  grade = 'F'
else:
  grade = 'U'
print(grade)
```
